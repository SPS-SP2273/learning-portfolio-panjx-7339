| Term       | Brief description                                                                                                                        |
|------------|------------------------------------------------------------------------------------------------------------------------------------------|
| Subsetting | selecting a specific portion of a sequence (e.g. lists, numpy arrays) or specific elements, much like taking a "subset"                  |
| Indexing   | accessing elements or subsets of elements within the sequence using a numerical index                                                    |
| Slicing    | getting a subset of elements from a sequence, by specifiying the start, end and (optional) step size                                   |
| Masking    | filters data in a sequence (e.g. numpy array) using a Boolean mask. Only elements that fulfill the condition (are  `True`) will be kept. |
