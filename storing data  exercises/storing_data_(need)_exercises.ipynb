# Exercise 1

1. Two similarities between lists and arrays
- both can store a sequence of objects 
- both are mutable

2. Two differences between lists and arrays
- lists can store elements of different data types, while arrays are limited to one data type 
- "Adding" two lists concatenates them, while array addition is an element-wise operation
