# 1 Lists, Arrays & Dictionaries

## 1.1 Let’s compare

```python
# Python lists
py_super_names = ["Black Widow", "Iron Man", "Doctor Strange"]
py_real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]
```

# 2 Lists vs. Arrays

## 2.1 Size

```python
py_list_2d = [[1, "A"], [2, "B"], [3, "C"], [4, "D"],
              [5, "E"], [6, "F"], [7, "G"], [8, "H"],
              [9, "I"], [10, "J"]]

np_array_2d = np.array(py_list_2d)      # Reusing the Python list 
                                        # to create a NEW
                                        # NumPy array
```

```python
print(len(py_list_2d))
print(len(np_array_2d))
print(np_array_2d.shape)
```

## 2.2 Arrays are fussy about type

Array elements must be of a single data type


```python
py_list = [1, 1.5, 'A']
np_array = np.array(py_list)
```

```python
py_list
```

```python
np_array # elements are converted to strings
```

## 2.3 Adding a number

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)         # Reusing the Python list
                                     # to create a NEW
                                     # NumPy array
```

```python
np_array + 10
```

```python
# py_list + 10        # Won't work!
```

## 2.4 Adding another list

```python
py_list_1 = [1, 2, 3, 4, 5]
py_list_2 = [10, 20, 30, 40, 50]

np_array_1 = np.array(py_list_1)
np_array_2 = np.array(py_list_2)
```

```python
py_list_1 + py_list_2
```

```python
np_array_1 + np_array_2 # array addition is an element-wise operation
```

```python
np.concatenate([np_array_1, np_array_2])
```

## 2.5 Multiplying by a Number

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)         
```

```python
py_list*2
```

```python
np_array*2
```

## 2.6 Squaring

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)
```

```python
np_array ** 2
```

```python
# py_list**2                      # Won't work!  
```

## 2.7 Asking questions

## 2.8 Mathematics

```python
py_list = [1, 2, 3, 4, 5]
np_array = np.array(py_list)         
```

```python
print(sum(py_list))    # sum() is a base Python function
print(max(py_list))    # max() is a base Python function
print(min(py_list))    # min() is a base Python function)
```

```python
print(np_array.sum())
print(np_array.max())
print(np_array.min())
print(np_array.mean())
print(np_array.std())
```
