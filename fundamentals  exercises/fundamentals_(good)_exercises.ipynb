```python
# Catches invalid and non-numerical values, and prompts the user to try again
def compute_grade():
    while True:
        try:
            score = int(input("Enter the student's score: "))
            if 0 <= score <= 100:
                break
            else:
                print("Invalid input! Please enter a score between 0 and 100.")
        except ValueError:
            print("Invalid input! Please enter a valid numerical score.")
    
    if 70 <= score <= 100:
        grade = "A"
    elif 50 <= score <= 69:
        grade = "B"
    elif 35 <= score <= 49:
        grade = "C"
    else:
        grade = "Fail"
    
    return f"The student's letter grade is: {grade}"

print(compute_grade())
```
