# Exercise 3

|  # | Term         | Description                                                                                                                    |
|:--:|--------------|--------------------------------------------------------------------------------------------------------------------------------|
|  1 | Function     | Blocks of code that are only executed when called, and contain reusable pieces of code, thus promoting code organisation.            |
|  2 | Arguments    | Values passed to a function when the latter is called, providing input |
|  3 | Comments     | Preceded by hashes (#), comments explain the purpose of specific sections or lines in the code, enhancing readability.         |
|  4 | Indentations | Define blocks of code (e.g. loop structures, function definitions), ensuring proper grouping and readability of the code. |
|  5 | Packages     | Collections of modules, which define functions with related functionalities (e.g. math module)                                 |
|  6 | .            | Allows access to specific attributes/methods of an object                                        |
|  7 | :            | Used to indicate the start of indented code blocks                       |
|  8 | []           | Square brackets define list structures (mutable, ordered sequences of data)                                    |
|  9 | [[[]]]       | Nested square brackets allow the creation of multi-dimensional lists or matrices                                               |
| 10 | {}           | Curly brackets define dictionaries (unordered collections of data)                                                             |
