# 1 A simple plot

```python
import matplotlib.pyplot as plt
```

## 1.1 Letâ€™s look at some code.

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values', color='red', linestyle='dashed')  # plots y_1 values against x 
plt.xlabel('x-values')  # labels x-axis
plt.ylabel('y-values')  # labels y-axis 
plt.title('X vs Y')     # sets title of graph
plt.grid(alpha=0.25)    # adjusts transparency of grid lines (min=0, max=1)
plt.legend(loc='upper left')  # creates legend and sets location
```

### Things to note

1. Some abbreviations:
| Long form | Abbreviation |
|:---------:|:------------:|
|   color   |       c      |
| linestyle |      ls      |
| linewidth |      lw      |

```python
# These two lines of code produce the same result 
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)

plt.plot(x, y_1, c='red', ls='dashed', lw=2)
```

2. Jupyter is an interactive environment, so you will see an output even if you omit `plt.show()`. However, it is good practice to include this line anyway so your code will also work in non-interactive environments (for instance, when the script is run directly from the command line).

3. Plotting functions usually have default values for the styling parameters. Thus, we can keep our plot simple like this:

```python
plt.plot(x, y_1, y_2)
```

4. You can split the arguments into separate lines to improve readability.
Both of the following forms are acceptable.

```python
plt.plot(x, y_1, color='red', linestyle='dashed', linewidth=2)
```

```python
plt.plot(x, y_2, label='Y values',
           color='red', linestyle='dashed')
```

## 1.2 Adding another plot

To plot the data of `y_2` on the same graph as `y_1`, we can add another line with the plot commmand:

```python
plt.plot(x, y_2, label='Y2 values', color='blue')
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')  # this line was added
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

Let me add another plot, but this time I will also include $x$ and $y$ 
error bars for the points. The plotting command I need to use for this is called `errorbar()`.

```python
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75
plt.errorbar(x, y_3,                      # plots y_3 versus x with
             xerr=x_error, yerr=y_error,  # defined x and y error bar sizes 
             label="Y3 with errors",
             color="green")
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

# 2 Better with NumPy

Often, it is easier to use NumPy arrays instead of Python lists. 

```python
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

np_x = np.array(x)    # converts list x into a numpy array
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

One of the advantages of NumPy arrays is that they allow us to generate data-related mathematical functions easily. Here, we can reuse our previous code to plot $x^2$ and $sin(x)$

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

Note that our plot does not look good because $sin(x)$ lies between +- 1, but $x^2$ 
 has no such bounds. One way to fix this is to add another y-axis that shares the same x-axis.

## We need another axis!

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

### Things to note

```python
# np.linspace(start, stop, num=50) returns evenly spaced numbers over a specified interval
# helps to smooth the sin curve out
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2
```

# 4 A real example: Global Warming

## 4.1 Plotting data from files

```python
# Plots Earthâ€™s daily land temperature data from the Berkeley Earth website
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)  # small alpha value to soften color of the plot
plt.ylim([-8, 8])                  # makes plot range symmetrical in the y direction
plt.hlines(0, date[0], date[-1], linestyle='--',  # adds horizontal line (y=0), 
           colors='grey')                  # which starts at date[0] and ends at date[-1]
                                        
```

## 4.2 Styles

Unlike myself, most people (very sensibly) might just want a decent-looking plot without spending time customising it. To facilitate this Matplotlib offers some standard style templates. I am going to use the one called **fivethirtyeight**.

```python
plt.style.use('fivethirtyeight')  # must be first line

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
```
