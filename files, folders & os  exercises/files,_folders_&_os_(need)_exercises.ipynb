# Exercise 1

```python
months = {
'jan': '01', 'feb': '02', 'mar': '03',
'apr': '04', 'may': '05', 'jun': '06',
'jul': '07', 'aug': '08', 'sep': '09',
'oct': '10', 'nov': '11', 'dec': '12'
}
```

## Task 1 and 2

```python
def rename_my_file(old_file_name):
  if "_" in old_file_name: # checks if file has been renamed
    date, remaining = old_file_name.split("_")
    month, day = date.split("-")
    new_month = months[month]
    return "-".join([new_month, day, remaining])
  else:
    return old_file_name
```

```python
rename_my_file('oct-08_singapore.txt')
```

## Task 3

```python
files = ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']
```

```python
new_files = []
for old_file_name in files: # renames a list of file names
  new_file_name = rename_my_file(old_file_name)
  new_files.append(new_file_name)
  print(f"{old_file_name} ----> {new_file_name}") # shows progress

print(f"Renamed files: {new_files}")
```

## Task 4

```python
import os
import glob
import shutil
```

```python
txt_files = glob.glob('os-collaboration-exercise-data/*.txt')
print(txt_files)
```

```python
# renames all .txt file names in place 
for old_path in txt_files:
  old_name = os.path.split(old_path)[-1]
  new_path = os.path.join('os-collaboration-exercise-data/', rename_my_file(old_name))
  if not os.path.exists(new_path):    # checks if renamed file already exists
    os.rename(old_path, new_path)
  else:
    print(f"New file path '{new_path}' already exists. Skipping rename.")

    
```

## Task 5

```python
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']
```

```python
# creates folders for each city
for city in cities:
  path = os.path.join('os-collaboration-exercise-data', city)
  try:
    os.mkdir(path)
  except FileExistsError:    # checks if folder already exists
    print(f'{path} already exists; skipping creation.')
    
```

## Task 6 and 7

```python
def list_all_files(city):
  return glob.glob(f'os-collaboration-exercise-data/{city}/*')
```

```python
list_all_files('Cairo')
```

```python
glob.glob('os-collaboration-exercise-data/Cairo')
```

```python
# copies renamed files to the corresponding folders of all cities 
for city in cities:
  city_files = glob.glob(f'os-collaboration-exercise-data/*{city.lower()}.txt')
  city_folder = f'os-collaboration-exercise-data/{city}'
  for file in city_files:
    shutil.copy(file, city_folder)
```
