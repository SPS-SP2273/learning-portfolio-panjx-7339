# 1 The for iterator

```python
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]
```

```python
# This is a not a good way to print messages 
## Does not scale well, difficult to make changes, error-prone
name=real_names[0]
print(f"{name} is a Marvel superhero!")

name=real_names[1]
print(f"{name} is a Marvel superhero!")

name=real_names[2]
print(f"{name} is a Marvel superhero!")
```

## 1.1 for with a list

```python
for name in real_names:
    print(f"{name} is a Marvel superhero!")
```

## 1.2 for with enumerate

```python
super_names = ["Black Widow", "Iron Man", "Doctor Strange"]
real_names = ["Natasha Romanoff", "Tony Stark", "Stephen Strange"]
```

```python
for count, name in enumerate(real_names):
    print(f'{count}: {name} is a Marvel superhero!')
```

```python
for index, name in enumerate(real_names):
    superhero_name = super_names[index]
    print(f'{name} is {superhero_name}!')
```

```python
# modifies the enumerate function to start counting from 100
for count, name in enumerate(real_names, 100):
    print(f'{count}: {name} is a Marvel superhero!')
```

## 1.3 for with range

```python
for i in range(5):
    print(i)
```

```python
# specifies start and end values (non-inclusive)
for i in range(5, 10):
    print(i)
```

```python
# we can also adjust the step size 
for i in range(1, 10, 3):
    print(i)
```

# 2 while

A `while` loop keeps on runnning **while** a condition is True.<br><br>
Unlike a for loop, we do not need to know beforehand the number of iterations needed.

```python
number = 0

while number < 5:
    print(number)
    number += 1
```
